```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_1 = cv2.imread('input/1.jpg')
img_2 = cv2.imread('input/IMG_20220315_174022.jpg')
```

```python
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)[300:600, 650:1100])
plt.title('my picture')
plt.show()
```

```python
hsv = cv2.cvtColor(img_1, cv2.COLOR_BGR2HSV)
#Blue
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])
# Yellow
lower_yellow = np.array([20, 50, 50])
upper_yellow = np.array([50, 255, 255])
# Red
lower_red = np.array([3, 50, 50])
upper_red = np.array([10, 255, 255])
```

```python
mask = cv2.inRange(hsv, lower_blue, upper_blue)

result = cv2.bitwise_and(hsv, hsv, mask=mask)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB))
plt.title('Blue')
plt.show()
```

```python
mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

result = cv2.bitwise_and(hsv, hsv, mask=mask)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB))
plt.title('Yellow')
plt.show()
```

```python
mask = cv2.inRange(hsv, lower_red, upper_red)

result = cv2.bitwise_and(hsv, hsv, mask=mask)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB))
plt.title('Red')
plt.show()
```
