```python
class Cube():
    
    def __init__(self):
        """
        Creamos un tensor de 3x3x6 para almacenar los colores del cubo, los colores son representados con valores del 1 al 6.
        """
        self.tensor = torch.tensor([torch.full([3,3],1).tolist(),
                     torch.full([3,3],2).tolist(),
                     torch.full([3,3],3).tolist(),
                     torch.full([3,3],4).tolist(),
                     torch.full([3,3],5).tolist(),
                     torch.full([3,3],6).tolist()])

    def rotate_plane(self, plane, clockwise):
        """
        Almacena los valores de los bordes de una cara en 4 vectores distintos, luego asigna esos valores en 
        la posicion correspondiente luego de realizar el giro de la cara, dependiendo de si es en sentido del reloj
        o contrario.
        """
        if clockwise:
            aux1 = torch.clone(self.tensor[plane][:,0])
            aux2 = torch.clone(self.tensor[plane][2,:])
            aux3 = torch.clone(self.tensor[plane][:,2])
            aux4 = torch.clone(self.tensor[plane][0,:])
            self.tensor[plane][:,0] = aux2
            self.tensor[plane][2,:] = aux3.flipud()
            self.tensor[plane][:,2] = aux4
            self.tensor[plane][0,:] = aux1.flipud()
        
        else:
            aux1 = torch.clone(self.tensor[plane][:,0])
            aux2 = torch.clone(self.tensor[plane][2,:])
            aux3 = torch.clone(self.tensor[plane][:,2])
            aux4 = torch.clone(self.tensor[plane][0,:])
            self.tensor[plane][:,0] = aux4.flipud()
            self.tensor[plane][2,:] = aux1
            self.tensor[plane][:,2] = aux2.flipud()
            self.tensor[plane][0,:] = aux3
            
        
    def R_(self):
        """
        Almacenamos los valores de la linea lateral derecha de la cara frontal del cubo en un nuevo vector, luego 
        al vector original le asignamos los valores de la linea lateral derecha de la cara inferior y continuamos con este 
        proceso hasta llegar a la linea lateral derecha superior, a la cual se le asigna el valor que guardamos al principio
        """
        aux = torch.clone(self.tensor[0][:,2])
        self.tensor[0][:,2] = self.tensor[1][:,2]
        self.tensor[1][:,2] = self.tensor[2][:,2]
        self.tensor[2][:,2] = self.tensor[3][:,2]
        self.tensor[3][:,2] = aux
        self.rotate_plane(4, False)
     
    def R(self):
        """
        Sigue la misma logica que R_ pero en sentido contrario
        """
        aux = torch.clone(self.tensor[0][:,2])
        self.tensor[0][:,2] = self.tensor[3][:,2]
        self.tensor[3][:,2] = self.tensor[2][:,2]
        self.tensor[2][:,2] = self.tensor[1][:,2]
        self.tensor[1][:,2] = aux
        self.rotate_plane(4, True)
        
    def L(self):
        aux = torch.clone(self.tensor[0][:,0])
        self.tensor[0][:,0] = self.tensor[1][:,0]
        self.tensor[1][:,0] = self.tensor[2][:,0]
        self.tensor[2][:,0] = self.tensor[3][:,0]
        self.tensor[3][:,0] = aux
        self.rotate_plane(5, True)
    
    def L_(self):
        aux = torch.clone(self.tensor[0][:,0])
        self.tensor[0][:,0] = self.tensor[3][:,0]
        self.tensor[3][:,0] = self.tensor[2][:,0]
        self.tensor[2][:,0] = self.tensor[1][:,0]
        self.tensor[1][:,0] = aux
        self.rotate_plane(5, False)
    
    def U(self):
        aux = torch.clone(self.tensor[0][0,:])
        self.tensor[0][0,:] = self.tensor[4][0,:]
        self.tensor[4][0,:] = self.tensor[2][2,:].flipud()
        self.tensor[2][2,:] = self.tensor[5][0,:].flipud()
        self.tensor[5][0,:] = aux
        self.rotate_plane(1, True)
        
    def U_(self):
        aux = torch.clone(self.tensor[0][0,:])
        self.tensor[0][0,:] = self.tensor[5][0,:]
        self.tensor[5][0,:] = self.tensor[2][2,:].flipud()
        self.tensor[2][2,:] = self.tensor[4][0,:].flipud()
        self.tensor[4][0,:] = aux
        self.rotate_plane(1, False)
    
    def D(self):
        aux = torch.clone(self.tensor[0][2,:])
        self.tensor[0][2,:] = self.tensor[5][2,:]
        self.tensor[5][2,:] = self.tensor[2][0,:].flipud()
        self.tensor[2][0,:] = self.tensor[4][2,:].flipud()
        self.tensor[4][2,:] = aux
        self.rotate_plane(3, True)
        
    def D_(self):
        aux = torch.clone(self.tensor[0][2,:])
        self.tensor[0][2,:] = self.tensor[4][2,:]
        self.tensor[4][2,:] = self.tensor[2][0,:].flipud()
        self.tensor[2][0,:] = self.tensor[5][2,:].flipud()
        self.tensor[5][2,:] = aux
        self.rotate_plane(3, False)
        
    def B(self):
        aux = torch.clone(self.tensor[4][:,2])
        self.tensor[4][:,2] = self.tensor[3][2,:].flipud()
        self.tensor[3][2,:] = self.tensor[5][:,0]
        self.tensor[5][:,0] = self.tensor[1][0,:].flipud()
        self.tensor[1][0,:] = aux
        self.rotate_plane(2, True)
        
    def B_(self):
        aux = torch.clone(self.tensor[4][:,2])
        self.tensor[4][:,2] = self.tensor[1][0,:]
        self.tensor[1][0,:] = self.tensor[5][:,0].flipud()
        self.tensor[5][:,0] = self.tensor[3][2,:]
        self.tensor[3][2,:] = aux.flipud()
        self.rotate_plane(2, False)
    
    def F_(self):
        aux = torch.clone(self.tensor[4][:,0])
        self.tensor[4][:,0] = self.tensor[3][0,:]
        self.tensor[3][0,:] = self.tensor[5][:,2]
        self.tensor[5][:,2] = self.tensor[1][2,:]
        self.tensor[1][2,:] = aux
        self.rotate_plane(0, False)
        
    def F(self):
        aux = torch.clone(self.tensor[4][:,0])
        self.tensor[4][:,0] = self.tensor[1][2,:]
        self.tensor[1][2,:] = self.tensor[5][:,2]
        self.tensor[5][:,2] = self.tensor[3][0,:]
        self.tensor[3][0,:] = aux
        self.rotate_plane(0, True)
        
    def Scramble(self, num=50, get_path=False):
        """
        Genera movimientos aleatorios para mezclar el cubo, el numero default de movimientos es 50 y si get_path=True, regresa
        una lista con los movimientos que fueron tomados
        """
        path = []
        for x in range(num):
            move = rnd.choice((self.R, self.R_, self.L, self.L_, self.U, self.U_, self.B, self.B_, self.F, self.F_, self.D, self.D_))
            path.append(move.__name__)
            move()
        if get_path:
            return path
        else:
            return None
    
    def Restart(self):
        """
        Regresa al cubo al estado inicial (ordenado)
        """
        self.tensor = torch.tensor([torch.full([3,3],1).tolist(),
                     torch.full([3,3],2).tolist(),
                     torch.full([3,3],3).tolist(),
                     torch.full([3,3],4).tolist(),
                     torch.full([3,3],5).tolist(),
                     torch.full([3,3],6).tolist()])
        return self.tensor.flatten()
    
    def Sequence(self, moves=None):
        """
        Ejecuta una secuencia de movimientos dados en forma de lista de strings. Ejemplo: ['R','F_','B','U_']
        """
        if moves == None:
            print("Please insert a list of moves. Example: ['R','F_','B','U_']")
        else:
            for x in moves:
                exec(f"self.{x}()")
                
    def score(self):
        """
        Puntaje para entrenar a la IA. Cada color en su lugar correcto suma 1 punto, el puntaje mínimo es 0 y el máximo
        es 48
        """
        score = 0
        for x in self.tensor:
            for ind1, y in enumerate(x):
                for ind2, z in enumerate(y):
                    if ind1 != 1 or ind2 != 1:
                        if x[ind1][ind2] == x[1][1]:
                            score+=1
        return score
    
    def action(self):
        """
        Placeholder. Esta sección no hace nada por ahora
        """
        pass
    
    def get_next_states(self):
        """
        Devuelve un diccionario con los pares (movimiento elegido, estado del cubo luego del movimiento) 
        para cada movimiento posible, de esta obteniendo todos los posibles futuros estados del cubo 
        a partir de un estado inicial
        """
        actions = [self.R, self.R_, self.L, self.L_, self.U, self.U_, self.B, self.B_, self.F, self.F_, self.D, self.D_]
        next_states = {}
        for action in actions:
            initial_state = copy.deepcopy(self.tensor)
            action()
            next_states[action]=self.tensor.flatten().float()
            self.tensor = initial_state
            
        return next_states
    
    def step(self, step, max_steps):
        """
        Devuelve el puntaje actual del cubo y si el cubo fue resuelto devuelve True, si no, False. También devuelve True
        si el cubo no ha sido resuelto en el numero maximo de movimientos permitido.
        """
        complete = torch.tensor([torch.full([3,3],1).tolist(),
                     torch.full([3,3],2).tolist(),
                     torch.full([3,3],3).tolist(),
                     torch.full([3,3],4).tolist(),
                     torch.full([3,3],5).tolist(),
                     torch.full([3,3],6).tolist()])
        if step >= max_steps:
            final_score = self.score()
            done = True
        elif torch.all(torch.eq(self.tensor,complete)):
            final_score = self.score() + 100
            done = True
        else:
            final_score = self.score()
            done = False
        return final_score, done
```

```python
import torch.nn as nn

class DeepQNetwork(nn.Module):
    """
    Creamos una red neuronal de 7 capas. la primera capa es el estado (futuro) del cubo (un vector de 54 elementos), 
    luego 5 capas de 256 neuronas (escogidas arbitrariamente) y por ultimo la ultima capa es la recompensa esperada por ese estado.
    """
    def __init__(self):
        super(DeepQNetwork, self).__init__()

        self.conv1 = nn.Sequential(nn.Linear(54, 256), nn.LeakyReLU(inplace=True))
        self.conv2 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv3 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv4 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv5 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv6 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv7 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv8 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv9 = nn.Sequential(nn.Linear(256, 256), nn.LeakyReLU(inplace=True))
        self.conv10 = nn.Sequential(nn.Linear(256, 1))

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.conv5(x)
        x = self.conv6(x)
        x = self.conv7(x)
        x = self.conv8(x)
        x = self.conv9(x)
        x = self.conv10(x)
        
        return x
```

```python
env = Cube()
```

```python
class opt():
    """
    Hyperparámetros de la red neuronal
    """
    batch_size = 512
    lr = 3e-9
    gamma = 0.93
    # model = DeepQNetwork()
    model = torch.load('trained_models/meganet_2/rubikmeganet_2_11000')
    initial_epsilon=1
    final_epsilon=0.05
    num_decay_epochs=5000
    num_epochs=40000
    save_interval=500
    replay_memory_size=3000000
    target_step = 10
    max_steps = 15
    scheduler_start=5000
    scheduler_end=6500
    scheduler_gamma=0.2
    log_path="tensorboard/meganet_2_retrain"
    saved_path=f"trained_models/{log_path.split('/')[1]}"
```

```python
def train(opt):
    rnd.seed(3)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(123)
        
    else:
        torch.manual_seed(123)
    if os.path.isdir(opt.log_path):
        shutil.rmtree(opt.log_path)
        
    os.makedirs(opt.log_path)
    if not os.path.exists(f"trained_models/{opt.log_path.split('/')[1]}"):
        os.makedirs(f"trained_models/{opt.log_path.split('/')[1]}")
    
    writer = SummaryWriter(opt.log_path)
    
    model = opt.model
    target_model = copy.deepcopy(model)
    
    optimizer = torch.optim.Adam(model.parameters(), lr=opt.lr)
    criterion = nn.MSELoss()
    scheduler = torch.optim.lr_scheduler.MultiStepLR(optimizer, milestones=[5000,6000,7000,8000,9000], gamma=opt.scheduler_gamma)
    
    env.Restart()
    env.Scramble(num=randint(1,10))
    state = env.tensor.flatten().float()
    if torch.cuda.is_available():
        model.cuda()
        target_model.cuda()
        state = state.cuda()
    replay_memory = deque(maxlen=opt.replay_memory_size)
    
    epoch = 0
    for x in range(opt.num_epochs):
        step = 0
        while True:
            next_steps = env.get_next_states()
            epsilon = opt.final_epsilon + (max(opt.num_decay_epochs - epoch, 0) * (
                    opt.initial_epsilon - opt.final_epsilon) / opt.num_decay_epochs)
            u = random()
            random_action = u <= epsilon
            next_actions, next_states = zip(*next_steps.items())
            next_states = torch.stack(next_states)
            if torch.cuda.is_available():
                next_states = next_states.cuda()
            model.eval()
            with torch.no_grad():
                predictions = model(next_states)[:, 0]
            if random_action:
                index = randint(0, len(next_steps) - 1)
            else:
                index = torch.argmax(predictions).item()

            next_state = next_states[index, :]
            action = next_actions[index].__name__
            
            next_actions[index]()

            reward, done = env.step(step, opt.max_steps)
            step+=1

            if torch.cuda.is_available():
                next_state = next_state.cuda()
            replay_memory.append([state, reward, next_state, done])
            if done:
                # final_score = env.score
                # final_cleared_lines = env.lines
                env.Restart()
                env.Scramble(num=randint(1,10))
                state = env.tensor.flatten().float()
                if torch.cuda.is_available():
                    state = state.cuda()
                break
            else:
                state = next_state
                continue

        if len(replay_memory) < opt.replay_memory_size / 3333:
            continue
        epoch += 1
        if epoch > 2300:
            opt.batch_size += 1
            if epoch > 3500:
                opt.batch_size += 2
        batch = sample(replay_memory, min(len(replay_memory), opt.batch_size))
        state_batch, reward_batch, next_state_batch, done_batch = zip(*batch)
        state_batch = torch.stack(tuple(state for state in state_batch))
        reward_batch = torch.from_numpy(np.array(reward_batch, dtype=np.float32)[:, None])
        next_state_batch = torch.stack(tuple(state for state in next_state_batch))

        if torch.cuda.is_available():
            state_batch = state_batch.cuda()
            reward_batch = reward_batch.cuda()
            next_state_batch = next_state_batch.cuda()

        q_values = model(state_batch)

        with torch.no_grad():
            next_prediction_batch = target_model(next_state_batch)
        model.train()

        y_batch = torch.cat(
            tuple(reward if done else reward + opt.gamma * prediction for reward, done, prediction in
                  zip(reward_batch, done_batch, next_prediction_batch)))[:, None]


        optimizer.zero_grad()
        loss = criterion(q_values, y_batch)

        writer.add_scalar('Train/Loss', loss.item()
                          # np.array(tbscores[-50:]).mean()
                          , epoch - 1)

        loss.backward()
        optimizer.step()
        
        # if epoch > opt.scheduler_start and epoch < opt.scheduler_end:
        scheduler.step()

        final_score = reward

        writer.add_scalar('Train/Score', final_score
                          # np.array(tbscores[-50:]).mean()
                          , epoch - 1)
        # writer.add_scalar('Train/Cleared lines', final_cleared_lines
        #                   # np.array(tbcleared_lines[-50:]).mean()
        #                   , epoch - 1)

        if epoch % opt.target_step == 0:
            target_model.load_state_dict(model.state_dict())
            # target_model = copy.deepcopy(model)
            
        print("Epoch: {}/{}, Action: {}, Score: {}".format(
            epoch,
            opt.num_epochs,
            action,
            final_score))

        # tbscores.append(final_score)
        # tbcleared_lines.append(final_cleared_lines)

        

        if epoch > 0 and epoch % opt.save_interval == 0:
            torch.save(model, "{}/rubik{}_{}".format(opt.saved_path, opt.log_path.split('/')[1], epoch))

        torch.save(model, "{}/rubik".format(opt.saved_path))
```

```python
import torch.nn as nn

class DeepQNetwork(nn.Module):
    def __init__(self):
        super(DeepQNetwork, self).__init__()

        self.conv1 = nn.Sequential(nn.Linear(3*3*6, 128), nn.LeakyReLU(inplace=True))
        self.conv2 = nn.Sequential(nn.Linear(128, 128), nn.LeakyReLU(inplace=True))
        self.conv3 = nn.Sequential(nn.Linear(128, 1))

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)

        return x
```
